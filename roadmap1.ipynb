```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.feature_extraction.text import CountVectorizer
import csv
from sklearn.feature_selection import SelectKBest, chi2, mutual_info_classif, f_classif
```

```python
df = pd.read_csv('/home/berkoc/Dersler/Graduation Project/prep_2.csv')
```

```python
def tfidf(train, test):
    vectorizer = TfidfVectorizer(use_idf=True,
                                 analyzer='word',
                                 strip_accents='unicode')
    train_tfidf = vectorizer.fit_transform(train)
    test_tfidf = vectorizer.transform(test)
    feature_names = vectorizer.get_feature_names()
    print(feature_names)
    return train_tfidf, test_tfidf, feature_names 
```

```python
x = df['content']
y = df['label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.80, test_size=0.20, random_state=42)
```

```python
train = pd.concat([x_train, y_train], axis = 1)
test = pd.concat([x_test, y_test], axis = 1)
```

```python
tfidf_train, tfidf_test, feature_names= tfidf(x_train, x_test)
```

```python
print(tfidf_train.shape)
print(y_train.shape)
print(tfidf_test.shape)
print(len(feature_names))
```

```python
chi = SelectKBest(chi2, k=50)
```

```python
x_train_features = chi.fit_transform(tfidf_train, y_train)
x_test_features = chi.transform(tfidf_test)
```

```python
print(chi.get_support(indices=True))
```

```python
feature_names = [feature_names[i] for i in chi.get_support(indices=True)]
```

```python
print(feature_names)
```

```python
model = MultinomialNB()
model.fit(x_train_features, y_train)
expected = y_test
predicted = model.predict(x_test_features)
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```
